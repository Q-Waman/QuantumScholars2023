## 1. Qubit


La información clásica puede ser almacenada en un sistema físico en estados clásicos , que son estados bien determinados y finitos.

Ejemplos:

1. En un bit clásico los estados clásicos pueden ser 1 y 0
2. Un dado puede tener 6 estados por lo que  estados clásicos pueden ser 1, 2, 3,4, 5 y 6.
3. Un ventilador electrico puede tener por ejemplo 3 estados ALTO, BAJO o MEDIO

La información cuántica puede ser almacenada en 2 estados medibles (0 y 1)  y además puede encontrarse en una superposición de estos estados esto es denominado un bit cuántico es decir un qubit, esto puede ser respresentado usando la notación Brac-ket, usando un vector de estado $ |\Psi \rangle $.

$$ | \Psi \rangle  = \alpha | 0 \rangle + \beta | 1 \rangle$$

Donde $\alpha$ y $\beta$ son números complejos denominados amplitudes.

Además los vectores $|0\rangle $ y $|1\rangle $ pueden ser respresentados como vectores columna

$$ | 0 \rangle = 
\begin{pmatrix}
1 \\
0 
\end{pmatrix}
$$

$$ | 1 \rangle = 
\begin{pmatrix}
0 \\
1 
\end{pmatrix}
$$

Por lo tanto el vector $ |\Psi \rangle $ toma la forma

$$ | \Psi \rangle  = \begin{pmatrix}
\alpha \\
\beta 
\end{pmatrix}$$

La norma al cuadrado de las amplitudes representan la probabilidad de medir su respectivo estado

$$Probabilidad(| 0\rangle ) = | \alpha |^{2}$$

$$Probabilidad(| 1\rangle ) = | \beta |^{2}$$

Como la probabilidad de observar alguno de los estados es segura (100%) se debe cumplir 

$$ | \alpha |^{2} +  | \beta |^{2} = 1$$

Se dice por tanto que el estado $ |\Psi \rangle $ está normalizado

### 2. Vector de estado en Qiskit

En estos primeros ejemplos veremos como usar qiskit para construir un vector de estado, para ello primero debemos importar de qiskit las herramientas de visualización para convertir un array de python a latex y luego convertir este objeto en un vector de estado de Qiskit.

Importamos lo que vamos a necesitar

```python
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import Statevector 
import numpy as np
```

y construimos primero los kets como vectores columna

El ket del estado $| 0 \rangle $ por ejemplo

```python
ket0 = [[1],[0]]
```

```python
array_to_latex(ket0)
```

El ket del estado $| 1 \rangle $ es 

```python
ket1 = [[0],[1]]
array_to_latex(ket1)
```

El estado $$ | + \rangle = \frac{1}{\sqrt{2}} | 0 \rangle + \frac{1}{\sqrt{2}} | 1\rangle $$ puede ser representado entonces

```python
ket_plus = [[1/np.sqrt(2)], [1/np.sqrt(2)]]
array_to_latex(ket_plus)
```

El estado $$  | \Phi \rangle = \frac{1 +2i}{3} | 0 \rangle - \frac{2}{3} | 1\rangle $$ podemos escribirlo

```python
ket_new = [[(1+ 2j) /3], [-2/3]]
array_to_latex(ket_new)
```

Ahora convertiremos todos estas listas a vectores de estado de qiskit, con la función Statevector

y los escribiremos en el formalismo bracket con el método draw

```python
Ket_0 = Statevector(ket0)
Ket_0.draw('Latex')
```

```python
Ket_1 = Statevector(ket1)
Ket_1.draw('Latex')
```

```python
Ketplus = Statevector(ket_plus)
Ketplus.draw('Latex')
```

```python
ketnew = Statevector(ket_new)
ketnew.draw('Latex')
```

Podemos además averiguar si nuestros estados están normalizados con el atributo is_valid de la clase Statevector

```python
Ket_0.is_valid()
```

```python
Ket_1.is_valid()
```

```python
Ketplus.is_valid()
```

```python
ketnew.is_valid()
```

En cambio un no normalizado nos dará:


$$  | \Phi \rangle = \frac{2i}{3} | 0 \rangle - \frac{2}{3} | 1\rangle $$

```python
ket_no_normalizado = [[(2j) /3], [-2/3]]
array_to_latex(ket_no_normalizado)
```

```python
KetNoNormalizado = Statevector(ket_no_normalizado)
KetNoNormalizado.draw('Latex')
```

```python
KetNoNormalizado.is_valid()
```

Además podemos obtener las norma de las amplitudes de probabilidad de estos vectores de estado con el atributo probabilities, por ejemplo del primer ket que creamos

Para el ket $| 0 \rangle$

```python
probs_ket_0 = Ket_0.probabilities()
print('probs: {}'.format(probs_ket_0))
```

Para el ket $| 1 \rangle$

```python
probs_ket_plus = Ketplus.probabilities()
print('probs: {}'.format(probs_ket_plus))
```

### 3. Dirac Bra

Para cada vector columna existe $ |\Psi \rangle $ un vector fila $ \langle \Psi |$ el cual es el transpuesto conjugado del vector columna

$$ \langle \Psi |  = | \Psi \rangle ^{\dagger} $$ 

En la notación de Dirac estos vectores fila son denominados Bras.

Por ejemplo obtengamos el transpuesto conjugado del estado visto anteriormente $|\Phi \rangle$

$$  | \Phi \rangle = \frac{1 +2i}{3} | 0 \rangle - \frac{2}{3} | 1\rangle $$

Obtenemos

$$  \langle \Phi | = \frac{1  - 2i}{3}  \langle  0 | - \frac{2}{3} \langle  1 | $$

Estos Bras también pueden ser representados en qiskit, por ejemplo la transpuesta conjugada del vector $| 0 \rangle$ 

```python
Bra0 = [1,0]
array_to_latex(Bra0)
```

la transpuesta conjugada de $| 1 \rangle $

```python
Bra1 = [0,1]
array_to_latex(Bra1)
```

Para el último Ket

```python
Bra2 = [(1-2j)/3,-2/3]
array_to_latex(Bra2)
```

### 4. Producto Interno

Un bra en general puede representarse de la siguiente manera 

$$ \langle \Psi | = \alpha^{\ast} \langle 0 | + \beta^{\ast} \langle 1 | = ( \alpha^{\ast}, \beta^{\ast})$$

Esto nos permite definir el producto interno entre vectores de estado, el cual esta dado por:

$$\langle \Psi ||| \Psi \rangle = \langle \Psi | \Psi \rangle$$

Esto permite definir la ortogonalidad entre 2 dos vectores de estado 

$$\langle \Psi_{1} | \Psi_{2} \rangle = 0 $$

y por tanto la ortonomalidad
$$\langle \Psi_{i} | \Psi_{j} \rangle = \delta_{ij} $$

Las bases de un qubit son ortonomales, con esta condición es suficiente para definir el espacio de Hilbert de los qubits


Por tanto podemos llamar a :
$ |  0 \rangle$ y $ |  1 \rangle$  bases del qubit, ya que cumplen ambas condiciones

Y podemos verificarlo haciendo uso de numpy, con el método .dot() entre dos objetos de la clase operator en qiskit


```python
from qiskit.quantum_info.operators import Operator
```

```python
Operador_ket_0 = Operator(ket0)
Operador_ket_0
```

```python
Operador_Bra_1 = Operator(Bra1)
Operador_Bra_1
```

```python
producto_interno_10 = np.dot(Operador_Bra_1, Operador_ket_0)
array_to_latex(producto_interno_10)
```

Igualmente a la inversa podemos calcular el producto $ \langle 0 | 1 \rangle$

```python
Operador_Bra_0 = Operator(Bra0)
Operador_Bra_0
```

```python
Operador_ket_1 = Operator(ket1)
Operador_ket_1
```

```python
producto_interno_01 = np.dot(Operador_Bra_1, Operador_ket_0)
array_to_latex(producto_interno_01)
```

### 4. Producto Externo

De igual forma se puede definir otro producto el producto externo 

$$ | \Psi \rangle \langle \phi |  = \begin{pmatrix}
\alpha  \\
\beta
\end{pmatrix} \otimes (\delta^{\ast}, \epsilon^{\ast}) = \begin{pmatrix}
\alpha\delta^{\ast}  && \alpha\epsilon^{\ast}\\
\beta\delta^{\ast} && \beta\epsilon^{\ast}
\end{pmatrix}$$

El caso especial cuando $| \Psi \rangle \langle \Psi |$ se denomina operador de proyección y permite proyectar cualquier estado en el estado $|\Psi \rangle$

Por ejemplo proyectemos el qubit general en la base $|  0 \rangle$, usando el operador de proyección $|  0 \rangle \langle 0 |$

$$|  0 \rangle \langle 0 | \Psi \rangle = |  0 \rangle \langle 0 | ( \alpha | 0 \rangle + \beta | 1 \rangle ) = \alpha |  0 \rangle \langle 0 | 0 \rangle + \beta |  0 \rangle \langle 0 | 1 \rangle = \alpha |  0 \rangle $$

por ello $\alpha = \langle 0 | \Psi \rangle$ y $\beta = \langle 1 | \Psi \rangle$

Un operador proyección importante es la indentidad

$$ |  0 \rangle \langle 0 |  + |  1 \rangle \langle 1 | = I$$

Este se obtiene de :


$$ | 0 \rangle \langle 0 |  = \begin{pmatrix}
1  \\
0
\end{pmatrix} \otimes (1, 0) = \begin{pmatrix}
1 * 1  && 1*0\\
0 * 1 &&  0*0
\end{pmatrix}= \begin{pmatrix}
1   && 0\\
0  &&  0
\end{pmatrix}$$ 

$$ | 1 \rangle \langle 1 |  = \begin{pmatrix}
0  \\
1
\end{pmatrix} \otimes (0, 1) = \begin{pmatrix}
0 * 0  && 0*1\\
1 * 0 &&  1*1
\end{pmatrix}= \begin{pmatrix}
0   && 0\\
0  &&  1
\end{pmatrix}$$ 

$$| 0 \rangle \langle 0 | + | 1 \rangle \langle 1 | = \begin{pmatrix}
1   && 0\\
0  &&  1
\end{pmatrix} = I$$   

Todo esto lo podemos hacer también en qiskit, encontremos el producto externo con la clase operator de qiskit y el método .outer() de numpy 

$$ | 0 \rangle \langle 0 | $$ 

```python
Operador_Bra_0
Operador_ket_0
Producto_externo_00 = np.outer(Operador_ket_0, Operador_ket_0)
array_to_latex(Producto_externo_00)
```

$$ | 1 \rangle \langle 1 | $$

```python
Operador_Bra_1
Operador_ket_1
Producto_externo_11 = np.outer(Operador_Bra_1, Operador_ket_1)
array_to_latex(Producto_externo_11)
```

$$ | 0 \rangle \langle 0 | + | 1 \rangle \langle 1 |$$ 

```python
array_to_latex(Producto_externo_00 + Producto_externo_11)
```

## 5. Esfera de Bloch

Podemos representar cada qubit individual como un punto en una esfera conocidad como la esfera de bloch, así una representación equivalente al qubit general
        $$ | \Psi \rangle  = \alpha | 0 \rangle + \beta | 1 \rangle$$

Es :

$$ | \Psi \rangle  = \cos(\frac{\theta}{2}) | 0 \rangle + e^{i\phi} \sin(\frac{\theta}{2}) | 1 \rangle  $$ 


Esta nueva representación se obtiene de considerar los eigenvalores de el operador espin en cualquier dirección en coordenadas esfericas, así cualquier qubit individual puede ser representado en la esfera de bloch donde :
$$   0 \leq \theta \leq \pi, 0 \leq \phi \leq 2\pi $$ 

Esta esfera puede ser visualizada en qiskit, usando el método plot_bloch_vector para un solo qubit, en el caso de multiples qubits en un circuito existe otra función que veremos más adelante

Primero importamos la función de el modulo visualization de Qiskit

```python
from qiskit.visualization import plot_bloch_vector
```

ahora podemos observar el ket $ | 0 \rangle $, en la esfera de Bloch de la siguiente manera 

```python

plot_bloch_vector([1, 0, 0], coord_type='spherical')
```

Como podemos observar, en la esfera de Bloch, tiene tres ejes, los ejes son X, Y y Z , que son la forma de representar a las matrices de Pauli en computación cuántica, los vectores en cada extremo de cada eje no son antiparalelos , ya que no son vectores en el sentido clásico de  la palabra son vectores de estado los cuales son ortogonales en cada extremo, esos vectores ortogonales son bases para el espacio de hilbert del qubit, y además son eigenvectores de su rescpectiva matriz de Pauli.

la función plot_bloch_vector(), puede usar la representación esférica del vector de estado recibiendo una lista con coordenadas esféricas que identifiquen el estado, el primer elemento de la listo es la coordenada r, la segunda es $\theta$ y la tercera es $\phi$:

Las coordenadas esférica del estado $| 0 \rangle$ por ejemplo son : 1, 0, 0

```python
plot_bloch_vector([1, 0, 0], coord_type='spherical')
```

Las coordenadas esférica del estado $| 1 \rangle$ por ejemplo son : 1, $\pi$, 0

```python
plot_bloch_vector([1, np.pi, 0], coord_type='spherical')
```

Las coordenadas esférica del estado $| + \rangle$ por ejemplo son : 1, $\frac{\pi}{2}$, 0

```python
plot_bloch_vector([1, np.pi/2, 0], coord_type='spherical')
```

```python
plot_bloch_vector([[0, 0, 1],[0, 0, -1]])
```
